```python
#Show a plot of the total number of responses.
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(3,2))
sns.histplot(df['Response'])
plt.title("Total Number of Response")
plt.show()

```

```python
#Show a plot of the response rate by the sales channel. 
plt.figure(figsize=(4,3))
sns.countplot(x = "Response", hue='Sales Channel', data = df)
plt.title("Response by Sales Channel")
plt.show()

```

```python
#Show a plot of the response rate by the total claim amount.
plt.figure(figsize=(4,3))
sns.countplot(x = "Response", hue='Total Claim Amount', data = df)
plt.title("Response by Total Claim Amoun")
plt.show()

```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x='Response', y='Total Claim Amount', data=df) 
plt.title("Response by Total Claim Amount")
plt.show()
```

```python
#Show a plot of the response rate by income.
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='Income', hue='Response', kde=True, bins=20)
plt.title("Response rate by Income")
plt.show()

```
